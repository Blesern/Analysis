```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
data = pd.read_csv(r'C:\Users\Nick\.anaconda\canadian_immegration_data.csv')
```

```python
data.head()
```

```python
data.set_index('Country', inplace=True)
```

```python
years=list(map(str, range(1980, 2014)))
```

```python
afghanistan=data.loc['Afghanistan', years]
afghanistan.head()
```

```python
#Line Plot for Immegration from Afghanistan 1980 - 2013
afghanistan.index=afghanistan.index.map(int)
afghanistan.plot(kind='line')

plt.title('Immegration from Afghanistan')
plt.ylabel('Number of Immegrants')
plt.xlabel('Years')

plt.text(1987, 3500, 'Collapse of Afghan State')
plt.show()
```

```python
#Start of an Area Plot of Top 5 Contributors to Canadian Immegration
data.sort_values(by='Total', ascending = False, axis=0, inplace=True)
```

```python
data_top5=data.head(5)
data_top5=data_top5[years].transpose()
```

```python
data_top5.plot(kind='area', alpha=0.25, stacked=False, figsize=(20, 10))

plt.title('Immegration Trend of Top 5 Contributors')
plt.ylabel('Number of Immegrants')
plt.xlabel('Years')

plt.show()
```

```python
#Start of a Histogram of Immegration from Scandinavian Countries to Canada
data_scan=data.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
data_scan.head()
```

```python
count, bin_edges=np.histogram(data_scan, 15)
data_scan.plot(kind='hist', figsize=(10, 6), bins=15, alpha=0.5, xticks=bin_edges, color=('maroon', 'darkblue', 'seagreen'))

plt.title('Immegration from Scandinavia to Canada')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immegrants')

plt.show()
```

```python
#Start of a Bar Chart of Immegration from Haiti to Canada
data_haiti=data.loc['Haiti', years]
data_haiti.head()
```

```python
data_haiti.plot(kind='bar', figsize=(10, 6))

plt.title('Immegration of Haiti to Canada')
plt.ylabel('Number of Immegrants')
plt.xlabel('Years')

plt.show()
```

```python
#Start of a Horizontal Bar Graph of Top 15 Contributors to Canadian Immegration
data.sort_values(by='Total', ascending = False, inplace = True)
data_top15=data['Total'].head(15)
data_top15
```

```python
data_top15.plot(kind='barh', figsize=(10, 6))

plt.xlabel('Number of Immegrants')
plt.title('Top 15 Countries that Contributed to Canadas Immegration')

plt.show()
```
